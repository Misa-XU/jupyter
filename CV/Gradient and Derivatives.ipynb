### Properties of Partial Derivatives

#### We say $\lim_{(x, y) \to (a, b)} f(x, y) = L$, provide that 
- For any positive number $\epsilon$ there exists a positive number $\delta = \delta(\epsilon)$ such that $|f(x, y) - L| < \epsilon$ for any $(x, y)$ satisfies $0 < \sqrt{(x - a)^2 + (y - b)^2} < \delta$

#### First partial derivatives
$$f_1(x, y) = \lim_{h \to 0} \frac{f(x + h, y) - f(x, y)}{h}$$
$$f_2(x, y) = \lim_{h \to 0} \frac{f(x, y + h) - f(x, y)}{h}$$

#### Alternative notations
$\frac{\partial}{\partial x} f(x, y) = f_1(x, y) = D_1f(x, y)$

#### Higher-Order Derivatives
$$\frac{\partial^2}{\partial x \partial y} f(x, y) = f_{12}(x, y) = f_{x, y}(x, y)$$

#### Equality of mixed partials
Suppose we have a function $f(x, y)$. Let's make $$Q = f(a + h, b + k) - f(a + h, b) - f(a, b + k) + f(a, b)$$ and two single variable function $$\mu(x) = f(x, b + k) - f(x, b) \, , \, \nu(y) = f(a + h, y) - f(a, y)$$ Then we have $$Q = \mu(a + h) - \mu(a) = \nu(b + k) - \nu(b)$$ Let's analyze $\mu(x)$ first. According to the Mean-Value Theorem, we have $$Q = h\mu'(a + \theta_1 h) = h(f_1(x + \theta_1 h, b + k) - f_1(x + \theta_1 h, b))$$ Then to find the derivative respect $y$ in $f_1$, we have $$kf_{12}(x + \theta_1h, b + \theta_2k) = f_1(x + \theta_1 h, b + k) - f_1(x + \theta_1 h, b)$$ So we have $$Q = khf_{12}(x + \theta_1h, b + \theta_2k)$$ Similarly, apply mean-Value Theorem to $\nu(y)$,  $$Q = khf_{21}(x + \theta_4h, b + \theta_3k)$$. When $\theta's$ tend to zero. we could have $f_{12} = f_{21}$

#### Chain Rule
For a single variable function $f(g(x))$, it is
$$\frac{d}{dx} f(g(x)) = \frac{d}{d g(x)} f(g(x)) \frac{d}{dx}g(x)$$
For a multiple variables function, let's start with a simple case $f(\mu(t), \nu(t)) = g(t)$:
$$
\begin{equation}
\begin{split}
\frac{d}{dt} g(t) & = \lim_{h \to 0} \frac{g(t + h) - g(t)}{h} = \lim_{h \to 0} \frac{f(\mu(t + h), \nu(t + h)) - f(\mu(t), \nu(t))}{h} \\
& = \lim_{h \to 0} \frac{f(\mu(t + h), \nu(t + h)) - f(\mu(t), \nu(t + h))}{h} + \lim_{h \to 0} \frac{f(\mu(t), \nu(t + h)) - f(\mu(t), \nu(t))}{h} \\
& = f_1(\mu(t), \nu(t))\mu'(t) + f_2(\mu(t), \nu(t)) \nu'(t)
\end{split}
\end{equation}
$$
Using a Leibniz notation, it is:
$$\frac{dz}{dt} = \frac{\partial z}{\partial x} \frac{dx}{dt} + \frac{\partial z}{\partial y} \frac{dy}{dt}$$

However, a more common case is that each variable be replaced by two variables, such that 
$$z = f(x, y),\, x = \mu(s, t),\, y = \nu(s, t)$$
Then in Leibniz notation, it would be written as:
$$\frac{\partial z}{\partial s} = \frac{\partial z}{\partial x}\frac{\partial x}{\partial s} + \frac{\partial z}{\partial y}\frac{\partial y}{\partial s}$$
$$\frac{\partial z}{\partial t} = \frac{\partial z}{\partial x}\frac{\partial x}{\partial t} + \frac{\partial z}{\partial y}\frac{\partial y}{\partial t}$$
The two equations could be combined as a matrix equation
$$\begin{pmatrix}\frac{\partial z}{\partial s} \\ \frac{\partial z}{\partial t}\end{pmatrix} =  \begin{pmatrix}\frac{\partial x}{\partial s} & \frac{\partial y}{\partial s} \\ \frac{\partial x}{\partial t} & \frac{\partial y}{\partial t} \end{pmatrix}  \begin{pmatrix}\frac{\partial z}{\partial x} \\ \frac{\partial z}{\partial y}\end{pmatrix}$$

#### Tangent Planes and Normal Lines
For a surface $z = f(x, y)$, the tangent plane at the point $(a, b, f(a, b))$ intersects the vertical plane y = b in a straight line; the slope of the intersection is $f_1(a, b)$ so it's parallel to the vector $\mathbf{T}_1 = \mathbf{i} + f_1(a, b)\mathbf{k}$. Similarly,  $\mathbf{T}_2 = \mathbf{j} + f_2(a, b)\mathbf{k}$. Then the normal vector
$$\mathbf{n} = \mathbf{T}_2 \times \mathbf{T}_1 = f_1(a, b)\mathbf{i} + f_2(a, b)\mathbf{j} + \mathbf{k}$$
And the tangent plane satisfy the equation
$$f_1(a, b)(x - a) + f_2(a, b)(y - b) + 1\cdot(z - f(a, b)) = 0$$
Which means that
$$z = f(a, b) + f_1(a, b)(x - a) + f_2(a, b)(y - b)$$

#### Linear Approximation of $f(x, y)$
According to the equation of the tangent plane
$$L(x, y) = f(a, b) + f_1(a, b)(x - a) + f_2(a, b)(y - b)$$

#### Differentiable
A function $f(x, y)$ is differentiable at the point $(a, b)$ if
$$\lim_{(h, k) \to (0, 0)} \frac{f(a + h, b + k) - f(a, b) - hf_1(a, b) - kf_2(a, b)}{\sqrt{h^2 + k^2}} = 0$$

#### Differentials
for a function $z = f(x_1, ... , x_n)$, it's differential $dz$ or $df$ is
$$\begin{equation}
\begin{split}
dz = df & = \frac{\partial z}{\partial x_1} d x_1+ ... + \frac{\partial z}{\partial x_n} d x_n \\
 & = f_1(x_1, ..., x_n)dx_1 + ... + f_n(x_1, ..., x_n) dx_n
\end{split}\end{equation}$$
Then df is an approximation to the $\Delta f$ such that $\Delta f = f(x_1 + dx_1, ... , x_n + dx_n) - f(x_1, ..., x_n)$. And when $dx_i$'s tend to zero, $df = \Delta f$.
This is another way of looking at linearization

#### Jacobian Matrix -- A Function from n-Space to m-Space
Suppose we have a vector $\mathbf{f} = (f_1, ..., f_m)$ of m functions, each depending on n variables $(x_1, ..., x_n)$; Then $\mathbf{f}$ define a transformation from $\mathbb{R}^n \to \mathbb{R}^m$. Let $\mathbf{y} = \mathbf{f}(x)$ such that $y_1 = f_1(x_1, ..., x_n), ..., y_m = f_m(x_1, ..., x_n)$. Then $\frac{\partial y_i}{\partial x_j}$ compose a $m \times n$ matrix (make sure that the output is in $\mathbb{R}^n$) $D\mathbf{f(x)}$ called __Jcobian matrix__ such that:
$$ D\mathbf{f(x)} =  \begin{pmatrix}
\frac{\partial y_1}{\partial x_1} & ... & \frac{\partial y_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial y_m}{\partial x_1} & ... & \frac{\partial y_m}{\partial x_n}
\end{pmatrix}$$
Example: <br>
Let $\mathbf{y = f(x)}$ be a transform from $\mathbb{R}^n \text{ to } \mathbb{R}^m$ and another transform $\mathbf{z = g(y)}$ given $D\mathbf{g(y)}$. Then $\mathbf{z = g \circ f = g(f(x))}$. According to the chain rule:
$$D\mathbf{(g \circ f)(x)} = D\mathbf{g(f(x))} D\mathbf{f(x)}$$
And the difference 
$$d\mathbf{y} = [dy_1 ... dy_n]^T = \begin{pmatrix} \frac{\partial y_1}{\partial x_1} & ... & \frac{\partial y_1}{\partial x_n} \\ \vdots & \ddots & \vdots \\ \frac{\partial y_m}{\partial x_1} & ... & \frac{\partial y_m}{\partial x_n} \end{pmatrix}
\begin{pmatrix} dx_1 \\ \vdots \\ dx_n \end{pmatrix} = D\mathbf{f(x)} d\mathbf{x}
$$

#### Jacobian Determinant
Suppose we have two functions $F = F(x, y)$ and $G = G(x, y)$, then the Jacobian is the determinant
$$\frac{\partial(F, G)}{\partial(x, y)} = \begin{vmatrix} \frac{\partial F}{\partial x} & \frac{\partial F}{\partial y} \\ \frac{\partial G}{\partial x} & \frac{\partial G}{\partial y} \end{vmatrix} = \begin{vmatrix} F_1 & F_2 \\ G_1 & G_2 \end{vmatrix}$$

#### Gradient
Suppose we have a function $f(x, y)$, then we define the gradient vector by
$$\nabla f(x, y) = f_1(x, y)\mathbf{i} + f_2(x, y)\mathbf{j}$$
where $\nabla$ is a operator defined by
$$\nabla = \mathbf{i} \frac{\partial}{\partial x} + \mathbf{j} \frac{\partial}{\partial y}$$

- $\nabla f(a, b)$ is a normal vector to the level curve of f that passes through (a, b)
- __Directional Derivatives__ $$D_{\mathbf{u}}f(a, b) = \lim_{h \to 0+} \frac{f(a + hu, b + hv) - f(a, b)}{h} = \mathbf{u} \cdot \nabla f(a, b)$$
Which could be proved by chain rule

#### Taylor Series and Approximations
In single variable case
$$F(a + h) = F(a) + hF'(a) + \frac{F''(a)}{2!}h^2 + ... + \frac{F^{(n)}(a)}{n!}h^n + \frac{F^{(n + 1)}(X)}{(n + 1)!}h^{n +1}$$
Assume h = 1 and a = 0, then
$$F(1) = F(0) + F'(0) + \frac{F''(0)}{2!} + ... + \frac{F^{(n)}(0)}{n!} + \frac{F^{(n + 1)}(\theta)}{(n + 1)!}$$
In bivariate case, first, for $f(a + h, b + k)$, let's consider it as a single-variable formula by applying a t such that 
$F(t) = F(a + th, b + tk)$, where $0 \le t \le 1$.

Clearly, $F(0) = F(a, b)$ and $F(1) = F(a + h, b + k)$. And
$$\begin{align*}
F'(t)  & = hf_1(a + th, b + tk) + kf_2(a + th, b + tk) \\
F''(t) & = h^2f_{11}(a + th, b + tk) + 2hkf_{12}(a + th, b + tk) + k^2f_{22}(a + th, b + tk) \\
F'''(t) & = h^3f_{111} + 3h^2kf_{112} + 3hk^2f_{122} + k^3f_{222}
\end{align*}$$
Which obviously has a binomial pattern. So we can rewrite $F''(x)$ as 
$$F''(t) = (hD_1 + kD_2)^2 f(a + th, b + tk)$$
In paricular, $F^{(m)}(0) = (hD_1 + kD_2)^m f(a, b)$
So the taylor formula for $f(a + h, b + k)$ is 
$$f(a + h, b + k) = \sum_{m = 0}^n \frac{1}{m!}(hD_1 + kD_2)^m f(a, b) + R_n(h, k) $$
Where the remainder term is given by
$$R_n = \frac{1}{(n + 1)!} (hD_1 + kD_2)^{n + 1}f(a + \theta h, b + \theta k)$$

#### Get Deriv Kernels
`cv2.getDerivKernels(dx, dy, ksize[, kx[, ky[, normalize[, ktype]]]]) → kx, ky`
- kx – Output matrix of row filter coefficients. It has the type ktype .
- ky – Output matrix of column filter coefficients. It has the type ktype .
- dx – Derivative order in respect of x.
- dy – Derivative order in respect of y.
- ksize – Aperture size. It can be CV_SCHARR , 1, 3, 5, or 7.
- normalize – Flag indicating whether to normalize (scale down) the filter coefficients or not. Theoretically, the coefficients should have the denominator  =2^{ksize*2-dx-dy-2} . If you are going to filter floating-point images, you are likely to use the normalized kernels. But if you compute derivatives of an 8-bit image, store the results in a 16-bit image, and wish to preserve all the fractional bits, you may want to set normalize=false .
- ktype – Type of filter coefficients. It can be CV_32f or CV_64F.

#### Gradient of Image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
wall = cv2.imread('dataset/wall.jpg')
```

```python
# Add noises
var = np.var(wall)
noise = np.random.normal(0,np.sqrt(var) * 0.05 ,wall.shape)
noised = noise + wall
```

```python
kx, ky = cv2.getDerivKernels(1, 1, 3)
```

```python
gradientX = cv2.filter2D(wall, -1, kx)
gradientY = cv2.filter2D(wall, -1, ky.T)
```

```python
#original
f, axs = plt.subplots(1, 2, figsize = (15, 15))
axs[0].imshow(gradientX)
axs[1].imshow(gradientY)
```

```python
gradientX_n = cv2.filter2D(noised, -1, kx)
gradientY_n = cv2.filter2D(noised, -1, ky.T)
```

```python
# shown that derivative is extremely sensitive to noises
f, axs = plt.subplots(1, 2, figsize = (15, 15))
axs[0].imshow(gradientX_n)
axs[1].imshow(gradientY_n)
```

```python
#### So let's may blur it first
blurred = cv2.GaussianBlur(noised, (5, 5), 3)
gradientX_gb = cv2.filter2D(blurred, -1, kx)
gradientY_gb = cv2.filter2D(blurred, -1, ky.T)
f, axs = plt.subplots(1, 2, figsize = (15, 15))
axs[0].imshow(gradientX_gb)
axs[1].imshow(gradientY_gb)
```
