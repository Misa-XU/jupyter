```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X = 1. / (np.arange(1, 11) + np.arange(0, 10)[:, np.newaxis]) # a good example of broadcast
y = np.ones(10)

n_alphas = 100
alphas = np.logspace(-10, -2, n_alphas)

coefs = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)
```

```python
ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
display(ax)
```
